```python
import sqlite3
import pandas as pd
import pprint

# Create connection
```

# 1. Database connection path

```python
db_path = '../data/raw/olist.sqlite'
```

# Create connection

```python
conn = sqlite3.connect(db_path)
```

# 2. Get all table names in the database

```python
query_tables = "SELECT name FROM sqlite_master WHERE type='table';"
tables_df = pd.read_sql_query(query_tables, conn)
tables_list = tables_df['name'].tolist()
```

# 3. Explore 'customers' table

```python
df_customers = pd.read_sql_query("SELECT * FROM customers", conn)
customers_dtypes = df_customers.dtypes.astype(str).to_dict()
customers_nulls = df_customers.isnull().sum().to_dict()
```

# 4. Explore 'orders' table

```python
df_orders = pd.read_sql_query("SELECT * FROM orders", conn)
orders_dtypes = df_orders.dtypes.astype(str).to_dict()
orders_nulls = df_orders.isnull().sum().to_dict()
order_status_counts = df_orders['order_status'].value_counts().to_dict()
```

# 5. Close connection

```python
conn.close()
```

# 6. Compile results into a single dictionary

```python
eda_results = {
    "tables_in_database": tables_list,
    "customers_table_summary": {
        "data_types": customers_dtypes,
        "missing_values_count": customers_nulls
    },
    "orders_table_summary": {
        "data_types": orders_dtypes,
        "missing_values_count": orders_nulls,
        "order_status_distribution": order_status_counts
    }
}
```

# 7. Print the dictionary cleanly so it can be copied

```python
print("--- Copy the dictionary below ---")
pprint.pprint(eda_results, sort_dicts=False)
```
